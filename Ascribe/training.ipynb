```python
import json
import pandas as pd
import openai
import numpy as np
import re
import language_tool_python
from torchtext.data.utils import get_tokenizer
with open("C:\\Ascribe\\app.config") as settings_file:
   appSettings = json.load(settings_file)
df = pd.read_csv(appSettings["training_data_path"])
to_drop = ['title','url','authors','timestamp','tags']
df.drop(to_drop, inplace=True, axis=1)
articles = np.array(df.values, dtype=object)
```

```python
TONE_OF_VOICE = "professional"
API_KEY = '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'
MODEL = "gpt-3.5-turbo"
```

```python
openai.api_key = API_KEY
tokenizer = get_tokenizer("basic_english")
tool = language_tool_python.LanguageTool('en-UK')
```

```python
import os
os.environ['OPENAI_API_KEY'] = API_KEY
```

```python
#f = open("f.jsonl", "a")
#TRAINING SET
f = open("C:\\PythonData\\f.jsonl", "a")
trainingJson = ""
x = 190
for i in range(155,x):
    print(i)

    noUrlArticle = re.sub(r'http\S+', '', str(articles[i][0]))
    tokenizedArticle = tokenizer(noUrlArticle)
    cleanedPrompt = ' '.join(tokenizedArticle)
    

    if len(tokenizedArticle) > 400:
        trainingJson +='{'
        trainingJson +='"prompt": ' + json.dumps(cleanedPrompt) + ','
        fullPrompt = cleanedPrompt
        print(str(i) + " is valid")
        completion = openai.ChatCompletion.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": f"Your role is to read any article the user passes in, and then convert it to a {TONE_OF_VOICE} tone of voice"},
            {"role": "user", "content": fullPrompt}
        ]
        )
        #completion.choices[1].content = cleanedPrompt
        cleanedCompletion = tool.correct(completion.choices[0].message.content)
        trainingJson +='"completion": '+ json.dumps(" " + cleanedCompletion + " END") +''
        if i == x-1:
            trainingJson +='}'
        else:
           trainingJson +='}\n' 
    
    

f.write(trainingJson)
f.close()
```
